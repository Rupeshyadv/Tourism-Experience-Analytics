```python
# merge all dataframes into one dataframe
df = transac_df.copy()

df = df.merge(user_df, on="UserId", how='left')
df = df.merge(city_df, on="CityId", how="left")

# Keep user country, drop city country, because both the df have CountryId.
df = df.rename(columns={"CountryId_x": "CountryId"})
df = df.drop(columns=["CountryId_y"])

df = df.merge(country_df, on="CountryId", how="left")

# Keep user region, drop country region, because both the df have RegionId.
df = df.rename(columns={"RegionId_x": "RegionId"})
df = df.drop(columns=["RegionId_y"])

df = df.merge(region_df, on="RegionId", how="left")

# Keep user continent, drop region continent, because both the df have ContinentId.
df = df.rename(columns={"ContinentId_x": "ContinentId"})
df = df.drop(columns=["ContinentId_y"])

df = df.merge(continent_df, on="ContinentId", how="left")

df = df.merge(updated_item_df, on="AttractionId", how="left")
df = df.merge(type_df, on="AttractionTypeId", how="left")
# there is no link between city and attraction in the given dataset, so we can drop the column AttractionCityId
df = df.drop(columns=["AttractionCityId"])

# rename the VisitMode to VisitModeId to avoid confusion with VisitModeId in mode_df
df = df.rename(columns={"VisitMode": "VisitModeId"})

df = df.merge(mode_df, on="VisitModeId", how="left")

df.head()
```

```python
df.isna().sum().sort_values(ascending=False)
df = df.dropna(subset=["CityId", "CityName"])
```

```python
# rating distribution
sns.histplot(df["Rating"], bins=10, kde=True)
plt.title("Distribution of Ratings")
plt.show()
```

```python
# user visit month distribution
sns.countplot(x=df["VisitMonth"])
plt.title("Distribution of User Visit Month")
plt.show()
```

```python
# country visits distribution
plt.figure(figsize=(10, 10))
country_counts = df["Country"].value_counts().sort_values(ascending=False).head(15) 
plt.pie(country_counts, labels=country_counts.index, autopct="%1.1f%%")
plt.title("Distribution of User Visit Country")
plt.show()
```

```python
# which cities are most visited
plt.figure(figsize=(10, 10))
city_counts = df["CityName"].value_counts().sort_values(ascending=False).head(15)
plt.barh(city_counts.index, city_counts.values)
plt.xlabel("Number of Visits")
plt.title("Top 15 Most Visited Cities")
plt.show()
```

```python
# avergae rating by city
avg_city_rating = (
    df.groupby("CityName")["Rating"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

avg_city_rating.plot(kind="bar")
plt.title("Top 10 Cities by Average Rating")
plt.ylabel("Average Rating")
plt.show()
```

```python
# which visit mode is most popular
visit_mode_counts = df["VisitMode"].value_counts()

visit_mode_counts.plot(kind="bar")
plt.title("Visit Mode Distribution")
plt.ylabel("Count")
plt.show()
```

```python
num_cols = df.select_dtypes(include="int64")
sns.heatmap(num_cols.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# which attraction is most visited
plt.figure(figsize=(10, 10))
attraction_counts = df["Attraction"].value_counts().sort_values().head(10)
plt.barh(attraction_counts.index, attraction_counts.values)
plt.xlabel("Number of Visits")
plt.title("Top 10 Most Visited Attractions")
plt.show()
```

```python
# which attractionType is most visited
plt.figure(figsize=(10, 10))
attraction_type_counts = df["AttractionType"].value_counts().sort_values().head(10)
plt.barh(attraction_type_counts.index, attraction_type_counts.values)
plt.xlabel("Number of Visits")
plt.title("Top 10 Most Visited Attraction Types")
plt.show()
```

```python
# rating vs visit mode
sns.boxplot(x="VisitMode", y="Rating", data=df)
plt.title("Ratings by Visit Mode")
plt.show()
```

```python
# avg rating by attraction type
avg_type = df.groupby("AttractionType")["Rating"].mean().reset_index().head(10)

plt.figure(figsize=(10, 3))
sns.barplot(x="AttractionType", y="Rating", data=avg_type)
plt.xticks(rotation=45)
plt.title("Average Rating by Attraction Type")
plt.show()
```

```python
# rating vs time of visit
sns.scatterplot(x="VisitMonth", y="Rating", data=df)
plt.title("Rating vs Visit Month")
plt.show()
```
